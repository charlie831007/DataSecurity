```python
# 隨機從dataset中挑選N筆資料
from os import walk
from os.path import join
from os import stat
import random
import hashlib
import time

mypath = "D:\Download\chunkhash\enron"

# store all the fullpath of files in the root directory
file_names = []
# store 1000 fullpath to upload
upload_files = []
# store 200 fullpath to check dulplicate
check_files = []

# three lists simulate cloud server
dirty_chunk = []
cloud_chunk = []
cloud_hash = []

# check file data
check_chunk = []
check_hash = []

for root, dirs, files in walk(mypath):
    for f in files:
        fullpath = join(root, f)
        file_names.append(fullpath)

# os.stat(file).st_size return size of the file in bytes
for i in range(10):
    random_int = random.randint(0, len(file_names) - 1)
    upload_files.append(file_names[random_int])
    
for i in range(2):
    random_int = random.randint(0, len(file_names) - 1)
    check_files.append(file_names[random_int])
```

```python
for i in range(len(upload_files)):
    print('file path: ' + upload_files[i])
    print('file size: ' + str(stat(upload_files[i]).st_size) + ' bytes')
```

```python
for i in range(len(check_files)):
    print('file path: ' + check_files[i])
    print('file size: ' + str(stat(check_files[i]).st_size) + ' bytes')
```

```python
chunk_size = 128

cloud_chunk = []
cloud_hash = []

for file in upload_files:
    with open(file, 'rb') as f:
        text = f.read()
    f.close()

    print(len(text))
    print(text)
    print()

    if(len(text) % chunk_size != 0):
        for i in range(chunk_size - (len(text) % chunk_size)):
            text += bytes.fromhex('30')

    print(len(text))
    print(text)
    print()

    text_chunk = []
    for i in range(0, len(text), chunk_size):
        text_chunk.append(text[i: i + chunk_size])
    cloud_chunk.extend(text_chunk)

    print(len(text_chunk))
    for chunk in text_chunk:
        print(chunk)
    print()

    text_hash = []
    for i in range(len(text_chunk)):
        hash_text = hashlib.sha256(text_chunk[i]).hexdigest()
        text_hash.append(hash_text)
    cloud_hash.extend(text_hash)

    print(len(text_hash))
    for th in text_hash:
        print(th)
    print()

print('cloud chunk', len(cloud_chunk))
for cc in cloud_chunk:
    print(cc)
print()

print('cloud hash', len(cloud_hash))
for ch in cloud_hash:
    print(ch)
```

```python
# 隨機從dataset中挑選N筆資料
from os import walk
from os.path import join
from os import stat
import random
import hashlib
import time

mypath = "D:\Download\enron"

# store all the fullpath of files in the root directory
file_names = []

# only store files size bigger than 5K
for root, dirs, files in walk(mypath):
    for f in files:
        fullpath = join(root, f)
        if(stat(fullpath).st_size >= 5000):
            file_names.append(fullpath)
```
